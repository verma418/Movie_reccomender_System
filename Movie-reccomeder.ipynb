```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
#Merging Both the Dataset on the basis of title (Movies and Credits dataset)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#making new dataset named movies having movie_id,genres,title,cast,crew,overview,keywords.
```

```python
movies = movies[['movie_id','title','genres','cast','crew','overview','keywords']]
```

```python
import ast 
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L               
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L               
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def convert4(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L               
```

```python
movies['crew']=movies['crew'].apply(convert4)
```
